```python
#Feature engineering

data['Order Date'] = pd.to_datetime(data['Order Date'],dayfirst = True)
data['Month'] = data['Order Date'].dt.month
data['Year'] = data['Order Date'].dt.year
```

```python
data.head()
```

```python
#Pivot Table

#pivot table representing Category's with maximum orders.
data.pivot_table(values = 'Order ID', index = 'Category', aggfunc = 'count')

#pivot table representing regions with maximum orders.
data.pivot_table(values = 'Order ID', index = 'Region' , aggfunc = 'count')

#pivot table representing span of years and sales.
data.pivot_table(values = 'Ship Date', index = 'Year', columns = 'Category', aggfunc='count')
```

```python
data1=data.copy()
```

```python
data1.head()
```

```python
data1.drop([
    'Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name',
    'Segment', 'City', 'State', 'Postal Code', 'Product ID', 'Sub-Category', 'Discount'
], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select columns to encode
cols_to_encode = ['Region', 'Category', 'Product Name','Country']

# Extract the categorical part
categorical_data = data1[cols_to_encode]

# Initialize encoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fit and transform
encoded_array = encoder.fit_transform(categorical_data)

# Create a DataFrame with proper column names
encoded_cols = encoder.get_feature_names_out(cols_to_encode)
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols, index=data1.index)

# Drop original categorical columns from data1
data1.drop(columns=cols_to_encode, inplace=True)

# Concatenate the encoded columns back to data1
data1 = pd.concat([data1, encoded_df], axis=1)
```
